# Required Imports

```python
import pandas as pd
import numpy as np
import plotly.express as px
```

# Get Data

```python
DF = pd.read_csv('time_series_19-covid-Confirmed.csv')
```

# Match Empty Region with Country Name

```python
DF.loc[DF["Province/State"].isna(), 'Province/State'] = DF.loc[DF["Province/State"].isna(), 'Country/Region']
```

# Function Below

```python
def bubbleMaps(df):
    newDF = df.copy()
    dates = newDF.columns[4:]
    df2 = pd.DataFrame(columns=['Place', 'Lat', 'Long', 'Date', 'Size', 'Cases', 'Text', 'Color'])
    #print(df2)
    for place in newDF["Province/State"].unique():
        for date in dates:
            Lat = df.loc[df["Province/State"] == place, "Lat"]
            Long = df.loc[df["Province/State"] == place, "Long"]
            number = df.loc[df["Province/State"] == place, date].item()
            text = "Number of cases: " + str(df.loc[df["Province/State"] == place, date].item()) + " in " +  str(place)
            case = int(df.loc[df["Province/State"] == place, date].item())
            size = (np.log(int(df.loc[df["Province/State"] == place, date].item()))/np.log(1e15))*500
            if (np.isinf(size)):
                size = 0
            if (number<3):
                color = "< 3"
            elif (3 <= number < 11):
                color = "< 11"
            elif (11 <= number < 21):
                color = "< 21"
            elif (21 <= number < 51):
                color = "< 51"
            elif (number > 50):
                color = "50+"
            temp = pd.DataFrame({"Place":[place], 
                             "Lat":int(Lat),
                         "Long":int(Long),
                         "Date": date,
                         "Size": int(size),
                         "Cases": case,
                         "Text": str(text),
                         "Color": color})
            df2 = df2.append(temp)
    fig = px.scatter_geo(df2, lat="Lat", lon="Long", color="Color",
                     hover_name="Text", size=(list(df2["Size"])),
                     animation_frame="Date",
                     text = "Text",
                     labels = {},
                     locationmode = 'country names')
    
    fig.update_geos(projection_type="natural earth",
               showcountries=True)
    fig.write_html('Graph.html')
    fig.show()
```

```python
bubbleMaps(DF)
```
